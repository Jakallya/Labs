```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans, AffinityPropagation, MeanShift, AgglomerativeClustering, DBSCAN
from sklearn.datasets import make_blobs
import os
```

```python
file_path = "C:\\Users\\Mi\\Downloads\\s1.txt"
ram_capacity = 2**20

if file_path.endswith('.txt') and os.path.isfile(file_path) and os.path.getsize(file_path) <= ram_capacity:
    data = np.loadtxt(file_path)
else:
    print("походу, такого файла нет или он слишком большой")
```

## KMeans

```python
kmeans = KMeans(n_clusters = 15, n_init = 10)
kmeans.fit(data)
labels = kmeans.predict(data)
kmeans_centroinds = kmeans.cluster_centers_
plt.figure(figsize = (20, 10))

plt.scatter(data[:, 0], data[:, 1], c=labels, cmap='jet', marker='o')
plt.plot(kmeans_centroinds[:, 0], kmeans_centroinds[:, 1], 'o', color='black')

plt.xlabel("X")
plt.ylabel("Y")
plt.title("Центроиды кластеров на основе модели KMeans")
plt.grid(True)

plt.show()
```

```python
affinity_propagation = AffinityPropagation(damping = 0.8, preference = -1000000000000)
affinity_propagation.fit(data)
affinity_clusters = affinity_propagation.predict(data)
affinity_centroids = data[affinity_propagation.cluster_centers_indices_]
plt.figure(figsize = (20, 10))

plt.plot(affinity_centroids[:, 0], affinity_centroids[:, 1], 'o', color='black')
plt.scatter(data[:, 0], data[:, 1], c=affinity_clusters, cmap='jet', marker='o')

plt.xlabel("X")
plt.ylabel("Y")
plt.title("Центроиды кластеров на основе модели AffinityPropagation")
plt.grid(True)

plt.show()
```

#### тут метод через матрицы(сходства, зависимости и доступности)

## MeanShift

```python
mean_shift = MeanShift(bandwidth = 100000)
mean_shift.fit(data)
mean_clusters = mean_shift.predict(data)
mean_centroids = mean_shift.cluster_centers_

plt.figure(figsize = (20, 10))

plt.plot(mean_centroids[:, 0], mean_centroids[:, 1], 'o', color='black')
plt.scatter(data[:, 0], data[:, 1], c=mean_clusters, cmap='jet', marker='o')

plt.xlabel("X")
plt.ylabel("Y")
plt.title("Центроиды кластеров на основе модели MeanShift")
plt.grid(True)

plt.show()
```

## AgglomerativeClustering

```python
agglomerative_clustering = AgglomerativeClustering(n_clusters=15)
agglomerative_clusters = agglomerative_clustering.fit_predict(data)

plt.figure(figsize = (20, 10))
plt.scatter(data[:, 0], data[:, 1], c=agglomerative_clusters, cmap='jet', marker='o')

plt.xlabel("X")
plt.ylabel("Y")
plt.title("Кластеры на основе модели AgglomerativeClustering")
plt.grid(True)

plt.show()
```

#### сначала все точки являются кластерами, потом они объединяются

## DBSCAN 

```python
dbscan = DBSCAN(eps=50000, min_samples=150)
dbscan.fit(data)
dbscan_clusters = dbscan.labels_

plt.figure(figsize = (20, 10))

plt.scatter(data[:, 0], data[:, 1], c=dbscan_clusters, cmap='jet', marker='o')

plt.xlabel("X")
plt.ylabel("Y")
plt.title("Центроиды кластеров на основе модели DBSCAN")
plt.grid(True)

plt.show()
```
